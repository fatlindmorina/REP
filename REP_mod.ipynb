```python
# Assuming you already have the 'Price' dataframe loaded
# Assuming the first column is 'Date' and the rest are asset prices
# Convert 'Date' column to datetime
df_price['Date'] = pd.to_datetime(df_price['Date'])

# Set 'Date' as index
df_price.set_index('Date', inplace=True)

# Calculate monthly returns for each asset
monthly_returns = df_price.pct_change()

#print("Monthly Returns:")
#print(monthly_returns)
monthly_returns
```

```python


# Assuming you already have the 'Price' dataframe loaded
# Assuming the first column is 'Date' and the rest are asset prices
# Convert 'Date' column to datetime
#df_price['Date'] = pd.to_datetime(df_price['Date'])

# Set 'Date' as index
#df_price.set_index('Date', inplace=True)

# Calculate logarithmic returns for each asset
log_returns = np.log(1 + df_price.pct_change())

#print("Logarithmic Returns:")
#print(log_returns)
log_returns
```

```python
# Select the period from 31.01.2004 until 29.02.2024
start_date = '2004-01-31'
end_date = '2024-02-29'
selected_returns = log_returns.loc[start_date:end_date]

# Calculate equal-weighted portfolio returns
portfolio_returns = selected_returns.mean(axis=1)

# Calculate cumulative returns
cumulative_returns = portfolio_returns.cumsum()

# Calculate portfolio standard deviation
portfolio_std = np.std(portfolio_returns)

# Print cumulative returns and portfolio standard deviation
print("Cumulative Portfolio Returns:")
print(cumulative_returns.iloc[-1])

print("\nPortfolio Standard Deviation:")
print(portfolio_std)
```
